```python
#train = np.loadtxt(fname="train.csv", delimiter=",", dtype="S")
#train = np.genfromtxt('train.csv', dtype= None)
file = pd.read_csv("train.csv") #Train Data Load

print(file.shape) #데이터의 기본 모양 확인  
file.head()
print(file.dtypes)
train = file.values
```

```python
def func(row):
    if row['Sex'] == 'male':
        row['Sex'] = 1
    elif row['Sex'] == 'female':
        row['Sex'] = 2
    if row['Embarked'] == 'C':
        row['Embarked'] = 1
    elif row['Embarked'] == 'Q':
        row['Embarked'] = 2
    elif row['Embarked'] == 'S':
        row['Embarked'] = 3
    else :
        return
    return row

# 제거 대상 Name, Age : 누락데이터가 생존이 너무 많음, Ticket, Cabin : 누락데이터 많음
# SibSp, Prarch는 모수가 다르기 때문에 비율로 데이터를 조정해서 뽑을 필요 있음.
data = file[["Pclass","Sex", "SibSp", "Parch", "Fare", "Embarked","Survived"]]
x_data = data.apply(func, axis=1).astype(np.float32).values
x_data = x_data[~np.isnan(x_data).any(axis=1)]
y_data = np.array(x_data)
y_data = y_data[:,[-1]]
x_data = x_data[:,0:6]
#y_data.append(file["Survived"])
print(x_data)
print(y_data)
```

```python
# placeholders for a tensor that will be always fed. 
X = tf.placeholder(tf.float32, shape=[None, 6]) 
Y = tf.placeholder(tf.float32, shape=[None, 1]) 


W = tf.Variable(tf.random_normal([6, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
z = tf.add(tf.matmul(X, W), b)
#hypothesis = tf.nn.sigmoid(z)
hypothesis = tf.nn.sigmoid_cross_entropy_with_logits(labels=Y,logits=z, name='hypothesis')
#hypothesis = tf.matmul(L1, W2) + b
#hypothesis = tf.nn.softmax(tf.add(tf.matmul(X, W), b))
#cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - tf.clip_by_value(hypothesis,1e-10,1.0)))
#cost1 = Y * tf.log(hypothesis)
#cost2 = (1 - Y) * tf.log(1 - tf.clip_by_value(hypothesis,1e-10,1.0))

#cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
cost = tf.reduce_mean(hypothesis, name='cost')



train = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(cost)
#train_steps = tf.train.AdamOptimizer(learning_rate).minimize(loss)


# Accuracy computation 
# True if hypothesis>0.5 else False 
predicted = tf.cast(tf.greater_equal(hypothesis, 0.5), tf.float32, name='pred') # 1 if >= 0.5
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32), name='acc') 

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    #ph, pc, pt = sess.run([hypothesis, cost, train], feed_dict={X: x_data, Y: y_data})
#    ph, pc1, pc2, pc, pt = sess.run([hypothesis, cost1, cost2, cost, train], feed_dict={X: x_data, Y: y_data})
    
#    print("\nHypothesis: ", ph)
#    print("\nCost: ", pc)
#    print("\nCost1: ", pc1)
#    print("\nCost2: ", pc2)
#    print("\nTrain: ", pt)

    
    for step in range(10001):
        cost_val, _ = sess.run([cost, train], feed_dict={X: x_data, Y: y_data})
        if step % 200 == 0: 
            print(step, cost_val) 


    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
    print("\nHypothesis: ", h, "\nCorrect (Y): ", c, "\nAccuracy: ", a)
```
